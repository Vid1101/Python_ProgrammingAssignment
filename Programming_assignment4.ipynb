```python
#Write a Python Program to Find the Factorial of a Number?
a = int(input("Enter a number"))
fact = 1
for i in range(1,a+1,1):
    fact = fact * i
    
print(fact)
```

```python
#Write a Python Program to Check Armstrong Number?
num = int(input("Enter the number to validate if it is armstrong or not"))
n = num
r,d = 0,0
while  n>0:
    d = n % 10
    r = r + d*d*d
    n = n//10
if r == num:
    print("%d is an armstrong number"%num)
else:
    print("%d is not armstrong number"%num)
    
```

```python
#Write a Python Program to Find Armstrong Number in an Interval?
num = int(input("Enter the interval"))
for i in range(num+1):
    n = i
    while n>0:
        d = n%10
        r = r+d*d*d
        n = n//10
        
    if r == i:
        print("{} is armstrong".format(i))
    else:
        print("{} is not armstrong".format(i))
    r,d = 0,0
```

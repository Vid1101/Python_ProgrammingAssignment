```python
#Q1 Write a Python Program to Check if a Number is Positive, Negative or Zero?
n = int(input("Enter a number"))
if n < 0:
    print("Negative Number", n)
elif n == 0:
    print("Its zero")
else:
    print("Its a positive number")
```

```python
#Q2 Write a Python Program to Check if a Number is Odd or Even?
n = int(input("Enter a number"))
if n%2 == 0:
    print("It's an even number")
else:
    print("It's an odd number")
```

```python
#Q3 Write a Python Program to Check Leap Year?
n = int(input("Enter the year to verify"))
if n%4 == 0:
    if (n//100)%4 == 0:
        print("Leap year")
    else:
        print("Not a leap year")
else:
    print("Not a leap year")
```

```python
#Q4 Write a Python Program to Check Prime Number?
from math import sqrt
n = int(input("Enter any number"))
prime_flag = 0

if(n > 1):
    for i in range(2, int(sqrt(n)) + 1):
        if (n % i == 0):
            prime_flag = 1
            break
    if (prime_flag == 0):
        print("Prime Number")
    else:
        print("Non Prime")
else:
    print("Non Prime")
```

```python
#Q5 Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?
from math import sqrt
for j in range(1, 10000):
    prime_flag = 0
    if(j > 1):
        for i in range(2, int(sqrt(j)) + 1):
            if (j % i == 0):
                prime_flag = 1
                break
        if (prime_flag == 0):
            print("Prime Number")
        else:
            print("Non Prime")
    elif j == 1 or j == 2:
        print("Prime Number")
    else :
        print("Non Prime")
    
```
